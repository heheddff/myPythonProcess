```python
def get_local(ip):
    api = "http://ip.taobao.com/service/getIpInfo.php?ip={}".format(ip)
    try:
        res = requests.get(api)
        #res.encoding = "utf-8"
        if res.status_code == 200:
            write(res.text)
        js = json.loads(res.text)
    except Exception as e:
        print(api,e)
        return False
    else:
        if js['code'] == 0:
            return js['data']
        else:
            print(js)
            return False

def write(mes):
    f = open('select.log',"a+")
    f.write(mes+"\n")
    f.close()
def wait(mes):
    f = open('waitip.log','a+')
    f.write(mes+"\n")
    f.close()
```

```python
ips = {}
with open("ip.log","r") as f:
    for line in f.read().split("\n"):   
            try:
                d = line.split("=")
                ip,num = d[0],int(d[1])
                if num > 1000:
                    ips[ip] = num
            except Exception as e:
                print(e)
                break
                print(line)
print(ips)
print(len(ips.keys()))
```
